```python
profile_file = "./config.share"
client = delta_sharing.SharingClient(profile_file)
```

```python
share_name = "core_share"
schema_name = "deltashare_core"

#table_url = profile_file + "#<share-name>.<schema-name>.<table-name>"
table_url = profile_file + f"#{share_name}.{schema_name}.messages"
```

```python
FILTER_DATE = "2026-03-03"

predicate_list = {
    "op": "equal",
    "children": [
    {"op": "column", "name":"StorageDateDayBR", "valueType":"string"},
    {"op":"literal","value":FILTER_DATE,"valueType":"string"}
    ]
} 

# predicate_list = {
#     "op": "and",
#     "children": [
#         {
#             "op": "equal",
#             "children": [
#                 {"op": "column", "name": "TenantId", "valueType": "string"},
#                 {"op": "literal", "value": "takeblip-eduardo-veloso", "valueType": "string"}
#             ]
#         },
#         {
#             "op": "equal",
#             "children": [
#                 {"op": "column", "name": "StorageDateDayBR", "valueType": "string"},
#                 {"op": "literal", "value": "2026-03-03", "valueType": "string"}
#             ]
#         }
#     ]
# }


# Transforma em string JSON
predicate_json = json.dumps(predicate_list)
```

```python
df = delta_sharing.load_as_pandas(table_url, jsonPredicateHints=predicate_json)
columns = ["OwnerIdentity", "TenantId", "StorageDateDayBR"]

print(f"Linhas retornadas: {len(df)}")
print(df[columns])
```
